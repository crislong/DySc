```python
import numpy as np
import myfunctions as mf
import class_rotationcurve as rc
import matplotlib.pyplot as plt
```

```python
radius = np.linspace(50,500,50)
rout = 105.
hr = 0.095

trial = mf.curve_total_model (radius/rout, 1/50 * (radius/rout), 1.12, 0.19, rout, hr, 100, 0.25)
dtrial = trial * 0 + 0.05 
trial1 = mf.curve_total_model (radius/rout, 1/50 * (radius/rout), 1.12, 0.,rout, hr, 100, 0.25)
for i in range(len(trial)):
    
    trial[i] += np.random.rand() * 0.1 - 0.1/2
```

```python
plt.plot(radius, trial, '.')
plt.plot(radius, trial1)

plt.yscale('log')
```

```python
rcurve = rc.rotationcurve(trial, dtrial, radius, 1/50 * radius, hr, 100, model=3)
```
